```python
import os
import pandas as pd
import numpy as np
import re
from sklearn.impute import SimpleImputer
```

```python
column_names = ['student_id','math_score','reading_score','writing_score']
df = pd.DataFrame(columns = column_names)

df = df.append({'student_id': 'ujlsq7296726iyzru', 'math_score': 75.0, 'reading_score': 69.0,'writing_score':68.0}, ignore_index=True)
df = df.append({'student_id': 'zdcku34151cfkek'}, ignore_index=True)
df = df.append({'student_id': 'anlya8548116kqokl', 'math_score': 72.0, 'reading_score': 80.0,'writing_score':75.0}, ignore_index=True)
df = df.append({'student_id': 'ykucw6122786vejtt', 'math_score': 79.0, 'reading_score': 78.0,'writing_score':77.0}, ignore_index=True)
```

```python
df
```

```python
df.dropna(axis=0, subset=['math_score','reading_score','writing_score'], thresh=2, inplace= True )
```

```python
df
```

```python
def numbers(element):    
    return "".join(filter(str.isnumeric, element))
```

```python
df['student_id'] = df['student_id'].apply(numbers)
df
```

```python
df.fillna(df['math_score'].median(), inplace = True)
df.fillna(df['reading_score'].median(), inplace = True)
df.fillna(df['writing_score'].median(), inplace = True)
```

```python
df['weighted_average'] = (df['math_score']*0.5  + df['reading_score']*0.2 +df['writing_score']*0.3)
```
